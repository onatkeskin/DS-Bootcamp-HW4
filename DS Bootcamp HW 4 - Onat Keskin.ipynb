```python
# Dummy Data
df = pd.DataFrame(np.random.randn(10, 4), 
                  columns=['col1', 'col2', 'col3', 'col4'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
df.head()
```

```python
df[['col1', 'col2']].plot(kind='bar');
```

```python
# plot with title

df_plot = df[['col1', 'col2']].plot(kind='bar');

df_plot.set_title('Dummy Data Plot of Column 1 and Column 2', fontsize=21,y=1);
```

```python
# legend lower-left corner.

df_plot = df[['col1', 'col2']].plot(kind='bar');

df_plot.set_title('Dummy Data Plot of Column 1 and Column 2', fontsize=21,y=1);

df_plot.legend(loc=3);
```

```python
# Horizontal bars
df_horizontal = df[['col1', 'col2']].plot.barh();

df_horizontal.set_title('Dummy Data Plot of Column 1 and Column 2', fontsize=21,y=1);

df_horizontal.legend(loc=3);
```

```python
# legend upper-right corner.

df_horizontal = df[['col1', 'col2']].plot.barh();

df_horizontal.set_title('Dummy Data Plot of Column 1 and Column 2', fontsize=21,y=1);

df_horizontal.legend(loc=1);
```
